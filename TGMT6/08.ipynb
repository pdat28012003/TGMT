```python
from PIL import Image
from imagetools import*
from IPython.display import display 
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Vẽ biểu đồ: Line Chart
x = [1, 2, 3, 4, 5]
y = [10, 16, 20, 30, 25]

x1 = [1, 2, 3, 4, 5]
y1 = [20, 15, 8, 9, 20]
x2 = [2,3,1,4,5]
y2 = [15,20,9,10,8]

plt.plot(x, y, 'bo-', label='Dữ liệu') # Vẽ đường 1
plt.plot(x1, y1, marker='o', color='r', linestyle='-', label='Dữ liệu-1')
plt.plot(x2, y2, marker='o', color='g', linestyle='-', label='Dữ liệu-2')

plt.legend()
plt.xlabel('Trục x') 
plt.ylabel('Trục y') 
plt.show() 
```

```python


categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 15, 22, 28]

plt.bar(categories, values, color='r', alpha=0.6) 
plt.xlabel('Danh mục') 
plt.ylabel('Giá trị') 
plt.show() 
```

```python

my_labels = ['A', 'B', 'C', 'D']
sizes = [30, 30, 30, 10]
my_colors = ['green', 'red', 'blue', 'yellow']

plt.pie(sizes, labels=my_labels, colors = my_colors, autopct='%1.1f%%') 
plt.title('Biểu đồ hình tròn')
plt.show()
```

```python

import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [15, 17, 28, 30, 35]

plt.scatter(x, y, c='violet', marker='o', label='Dữ liệu')  
plt.title('Biểu đồ phân tán ví dụ')  
plt.xlabel('Trục x')  
plt.ylabel('Trục y')  
plt.legend()  
plt.show()  
```

```python

import matplotlib.pyplot as plt

data = [15, 18, 25, 30, 35, 45, 60, 55, 65]

plt.boxplot(data)  
plt.title('Biểu đồ hộp ví dụ')  
plt.ylabel('Giá trị')  
plt.show()  
```

```python
import matplotlib.pyplot as plt

data = [15, 18, 22, 30, 35, 45, 50, 55, 65]

plt.violinplot(data) 
plt.title('Biểu đồ violin ví dụ')  
plt.ylabel('Giá trị')  
plt.show()  
```
